### Задание №2

```python
file = open(r'D:\cuda\lab2\lab2\text2.txt','r')
n_stream = int(file.readline().strip())
N = [int(i) for i in file.readline().strip().split(' ')]
arr1 = [float(i) for i in file.readline().strip().split(' ')]
arr2 = [float(i) for i in file.readline().strip().split(' ')]

for i in range(1,n_stream+1):
    plt.subplot(4,2,i)
    plt.plot(N,arr1[(i-1)*len(N):i*len(N)],label='one transactions')
    plt.plot(N,arr2[(i-1)*len(N):i*len(N)],label='many transactions')
    plt.title(str(i))
    plt.xscale('log')
    plt.xlabel('N')
    plt.ylabel('Time, ms')
plt.legend(loc='upper right')
```
